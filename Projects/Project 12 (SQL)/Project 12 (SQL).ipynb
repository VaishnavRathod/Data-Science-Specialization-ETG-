```python
import pandas as pd
import sqlite3 as sql
```

```python
df = pd.read_csv('IMDB_Movie_Data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(round(df['Revenue (Millions)'].mean(), 2))
df['Metascore'] = df['Metascore'].fillna(round(df['Metascore'].mean(), 2))
```

```python
df.isnull().sum()
```

```python
conn = sql.connect('Project 12 (SQL).db')
```

```python
df.to_sql('IMDB_Movie_Data', conn, if_exists='replace')
conn.close()
```
