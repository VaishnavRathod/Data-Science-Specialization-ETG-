# Data Analysis - COVID

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing dataset

df = pd.read_csv("country_vaccinations.csv")
df.head(10)
```

```python
df.isnull().sum()
```

## Project 5 

### Q1 Add year column in the dataset from the date column

```python
year = []
for i in range(len(df)):
    year.append(df["date"][i][0:4])
    
df.insert(loc= 3, column= "Year",value= year)
```

```python
df.head()
```

### Q2 Add month column in the dataset from the date column [1-12].

```python
month = []
for i in range(len(df)):
    month.append(df["date"][i][5:7])
    
df.insert(loc= 3, column= "Month",value= month)
```

```python
df.head()
```

### Q3 Add day column in the dataset from the date column.

```python
day = []
for i in range(len(df)):
    day.append(df["date"][i][8:])
    
df.insert(loc= 3, column= "Day",value= day)
```

```python
df.head()
```

### Adding week day column (Mon - Sun)

```python
week_day = [] 
for i in range(len(df)):
    week_day.append(pd.Timestamp(df["date"][i]).day_name())
    
df.insert(loc= 3, column= "Week Day",value= week_day)
```

```python
df.head()
```

## Project 6

### Q1 Add Month column in the datset from the date column [Jan - Dec]


```python
month_name = [] 
for i in range(len(df)):
    month_name.append(pd.Timestamp(df["date"][i]).month_name())

df = df.drop(axis = 1,columns="Month")    
df.insert(loc= 5, column= "Month",value= month_name)
```

```python
df.head()
```

## Project 7


### Q1 How many people are total vaccinated in India in May 2021?

```python
df = df.fillna(value=0)
```

```python
df.head()
```

```python
df_ind = df[df['country'] == "India"]
df_mon = df_ind[df_ind['Month'] == 'May']
df_y = df_mon[df_mon['Year'] == '2021']
df_y
```

```python
print("The total vaccinations of the people in May 2021 is: ")
print(int(df_y['total_vaccinations'][-1:]) - int(df_y['total_vaccinations'][:1]))
```

### Q2 How many people are vaccinated in Albania on 2022-01-12?

```python
df_alb = df[df['country'] == "Albania"]
df_date = df_alb[df_alb['date'] == '2022-01-12']
```

### Q1 Find the monthly number of total vaccinations from Jan to Dec in 2021.

```python
df_9 = df[df['Year'] == '2021'].groupby(by='Month').sum()['total_vaccinations']
```

```python
sort_order=['January','February','March','April','May','June','July','August','September','October','November','December']
for i in sort_order:
    print(i , "------" ,df_9[i])
```
