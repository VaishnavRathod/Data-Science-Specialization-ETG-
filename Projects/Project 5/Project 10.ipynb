```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Project 5 

### Q1 Add year column in the dataset from the date column

```python
year = []
for i in range(len(df)):
    year.append(df["date"][i][0:4])
    
df.insert(loc= 3, column= "Year",value= year)
```

```python
df.head()
```

### Q2 Add month column in the dataset from the date column [1-12].

```python
month = []
for i in range(len(df)):
    month.append(df["date"][i][5:7])
    
df.insert(loc= 3, column= "Month",value= month)
```

```python
df.head()
```

### Q3 Add day column in the dataset from the date column.

```python
day = []
for i in range(len(df)):
    day.append(df["date"][i][8:])
    
df.insert(loc= 3, column= "Day",value= day)
```

```python
df.head()
```

### Adding week day column (Mon - Sun)

```python
week_day = [] 
for i in range(len(df)):
    week_day.append(pd.Timestamp(df["date"][i]).day_name())
    
df.insert(loc= 3, column= "Week Day",value= week_day)
```

```python
df.head()
```

## Project 10

### Q1 Find the number of Total vaccinations in India in year 2020, 2021, and 2022.

```python
df_10 = df[df['country'] == 'India'].groupby(by='Year').sum()['total_vaccinations']
```

```python
years = [2020, 2021, 2022]
df_10_final = []
for i in years:
    if str(i) in df_10:
        df_10_final.append(int(df_10[str(i)]))
    else:
        df_10_final.append(0)
    
```

```python
for i in range(len(years)):
    print("The number of Total vaccinations in India in year", years[i] , "is", df_10_final[i])

plt.figure(figsize=(10,8))
plt.bar(range(len(years)), df_10_final, tick_label=years, data=df_10_final)
plt.title("The number of Total vaccinations in India in each year")
plt.xlabel("Years")
plt.ylabel("Number of vaccinations")
for x,y in enumerate(df_10_final):
    plt.text(x, y, str(y), fontweight = 'bold', horizontalalignment='center')
plt.show()
```

### Q2 Compare the number of Total vaccinations in year 2020 of India and USA.

```python
df_india['2020'] = 0
df_india
```

```python
count = [df_india['2020'], df_usa['2020']]
country = ['India','USA']
```

```python
plt.bar(country, count)
plt.title("Comparision of number of total vaccinations of India and USA in year 2020.", pad= 20)
for x,y in enumerate(count):
    plt.text(x, y, str(y), fontweight = 'bold', horizontalalignment='center')
plt.ylabel("Number of total vaccinations")
plt.show()
```

### Q4 Find the number of vaccination in each month in India in the year 2021.

```python
df_10_ = df[df['country'] == 'India']
```

```python
df_10_ = df[df['Year'] == '2021'].groupby(by='Month').sum()['total_vaccinations']
```

```python
df_10_month = {}
sort_order=['January','February','March','April','May','June','July','August','September','October','November','December']
for i in sort_order:
    df_10_month[i] = df_10_[i]
```

```python
df_10_month
```

### Q5 Which month has the most number of total vaccinations in India in 2021? 

```python
months = list(df_10_month.keys())
number = list(df_10_month.values())
plt.figure(figsize=(15,6))
plt.bar(months, number)
for x,y in enumerate(number):
    plt.text(x, y, str(y), size = "medium", horizontalalignment='center')
plt.title("The number of Total vaccinations in India in each month in the year 2020")
plt.xlabel("Months")
plt.ylabel("Number of vaccinations")
plt.show()
```
